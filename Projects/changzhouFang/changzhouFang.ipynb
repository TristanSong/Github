```python
# 来看看新北、天宁、钟楼、武进行政区分别到最近BRT、地铁、公园距离
import matplotlib.pyplot as plt
import pandas as pd

com_BRT = pd.read_csv('./小区_BRT.csv', sep=',', header=None)
cnames = ['community', 'distance', 'station', 'district']
com_BRT.columns = cnames
com_BRT.drop(com_BRT.index[com_BRT.district=='溧阳'], inplace=True)
com_BRT.drop(com_BRT.index[com_BRT.district=='金坛'], inplace=True)
com_BRT.drop(com_BRT.index[com_BRT.district=='戚墅堰'], inplace=True)
print('小区到BRT平均距离：%.fm'%com_BRT.distance.mean())

com_METRO = pd.read_csv('./小区_METRO.csv', sep=',', header=None)
cnames = ['community', 'distance', 'station', 'district']
com_METRO.columns = cnames
com_METRO.drop(com_METRO.index[com_METRO.district=='溧阳'], inplace=True)
com_METRO.drop(com_METRO.index[com_METRO.district=='金坛'], inplace=True)
com_METRO.drop(com_METRO.index[com_METRO.district=='戚墅堰'], inplace=True)
print('小区到METRO平均距离：%.fm'%com_METRO.distance.mean())

com_PARK = pd.read_csv('./小区_PARK.csv', sep=',', header=None)
cnames = ['community', 'distance', 'station', 'district']
com_PARK.columns = cnames
com_PARK.drop(com_PARK.index[com_PARK.district=='溧阳'], inplace=True)
com_PARK.drop(com_PARK.index[com_PARK.district=='金坛'], inplace=True)
com_PARK.drop(com_PARK.index[com_PARK.district=='戚墅堰'], inplace=True)
print('小区到PARK平均距离：%.fm'%com_PARK.distance.mean())
```
